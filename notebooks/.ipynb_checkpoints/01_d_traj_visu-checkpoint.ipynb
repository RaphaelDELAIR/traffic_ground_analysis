```python
def geoencode_groundtraffic(
    self,
    airport: str,
    dt_sec: int = 5,
):
    def timestamp(t):
        return pd.to_datetime(t).timestamp() * 1000

    # The following cast secures the typing
    _airport = airports[airport] if isinstance(airport, str) else airport
    if _airport is None or _airport.runways.shape.is_empty:
        return None

    source = (
        self.iterate_lazy()
        .inside_bbox(_airport)
        .onground()
        .resample(f"{dt_sec}s")
        .eval()
        .data[["timestamp", "latitude", "longitude", "flight_id"]]
    )

    slider = alt.binding_range(
        step=1000 * dt_sec,  # 1sec
        min=timestamp(min(source["timestamp"])),
        max=timestamp(max(source["timestamp"])),
    )

    select_date = alt.selection_single(
        fields=["timestamp"],
        bind=slider,
        init={"timestamp": timestamp(min(source["timestamp"]))},
        name="slider",
    )

    return _airport.geoencode() + alt.Chart(source).mark_circle(size=30).encode(
        latitude="latitude",
        longitude="longitude",
        color=alt.Color("flight_id"),
    ).add_selection(select_date).transform_filter(
        "(minutes(datum.timestamp) == minutes(slider.timestamp[0])) && "
        "(seconds(datum.timestamp) == seconds(slider.timestamp[0])) && "
        "(hours(datum.timestamp) == hours(slider.timestamp[0]))"
    )


import types
```

```python
t_dep_sample = t_dep.between(
    pd.Timestamp("2019-10-05 11:00:00+00:00"), pd.Timestamp("2019-10-05 12:00:00+00:00")
)
```

```python
t_dep_sample.geoencode_groundtraffic = types.MethodType(
    geoencode_groundtraffic, t_dep_sample
)
t_dep_sample.geoencode_groundtraffic("LSZH")
```

```python
f=t_dep["SWR786_14674"]
filter_dict=dict(
            compute_track_unwrapped=21, compute_track=21, compute_gs=21
        )
track_threshold=90

# --------------------------------
_airport = airports['LSZH']
within_airport = f.inside_bbox(_airport)
if within_airport is None:
    print(" None")

parking_position = within_airport.on_parking_position(_airport).next()
if parking_position is None:
    print(" None")

after_parking = within_airport.after(parking_position.start)
assert after_parking is not None

in_movement = after_parking.moving()

if in_movement is None:
    print(" None")

direction_change = (
    # trim the first few seconds to avoid annoying first spike
    in_movement.first("5T")
    .last("4T30s")
    .cumulative_distance()
    .unwrap(["compute_track"])
    .filter(**filter_dict)
    .diff("compute_track_unwrapped")
    .query(f"compute_track_unwrapped_diff.abs() > {track_threshold}")
)

if direction_change is None:
    print(" None")

```
