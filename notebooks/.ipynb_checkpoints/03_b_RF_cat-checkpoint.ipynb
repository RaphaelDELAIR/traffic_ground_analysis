```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import altair as alt

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt

from collections import namedtuple

import os


def within_Xmin(real, pred, m=1):
    assert len(real) == len(pred)
    c = 0
    for r, p in zip(real, pred):
        if abs(r - p) < m:
            c += 1
    return c / len(real) * 100


def plot_feature_importance(
    feature_columns,
    clf,  # saveas="../prediction_results/result.png"
):
    col = feature_columns
    # modelname.feature_importance_
    y = clf.feature_importances_
    # plot
    fig, ax = plt.subplots()
    width = 0.2  # the width of the bars
    ind = np.arange(len(y))  # the x locations for the groups
    ax.barh(ind, y, width, color="green")
    ax.set_yticks(ind + width / 10)
    ax.set_yticklabels(col, minor=False)
    plt.title("Feature importance in RandomForest Classifier")
    plt.xlabel("Relative importance")
    plt.ylabel("feature")
    plt.figure(figsize=(5, 8))
    fig.set_size_inches(6.5, 9.5, forward=True)
    return fig


def train_rf_cat(
    flight_df,
    feature_columns,
    target_column,
    query,
    dirname="../prediction_results/rf/",
):
    """
    function that will train and plot performance metric
    """
    test_name = dirname.split("/")[-2]

    flight_df = flight_df.query(query)

    data_df = flight_df[feature_columns + [target_column]].dropna()

    input_df = data_df[feature_columns]
    target_df = data_df[[target_column]]

    data_train, data_test, labels_train, labels_test = train_test_split(
        input_df,
        target_df,
        test_size=0.20,
    )

    clf = RandomForestClassifier(n_estimators=150, min_samples_split=10)

    # fit the model on the whole dataset
    clf.fit(data_train, labels_train)
    # make a single prediction
    predicted = clf.predict(data_test)

    feature_chart = plot_feature_importance(feature_columns, clf)
    res_txt = "Results for Random Forest on the following features :\n"
    res_txt += " ".join(feature_columns) + "\n"
    res_txt += f"Target var : {target_column}\n"
    res_txt += "Selecting the subset with following query : \n" + query + "\n"
    res_txt += f"n_sample training : {data_train.shape[0]}\nn_sample test : {data_test.shape[0]}\nn_sample_tot={input_df.shape[0]}\n"

    res_txt += classification_report(labels_test.values.reshape((-1,)), predicted)

    os.makedirs(os.path.dirname(dirname), exist_ok=True)

    fig = plot_confusion_matrix(
        clf,
        data_test,
        labels_test,
        # display_labels=["<8", ">=8"],
        cmap=plt.cm.Blues,
        normalize="pred",
    )
    plt.savefig(dirname + "conf_mat.png")

    with open(dirname + "res.txt", "w") as text_file:
        text_file.write(res_txt)

    feature_chart.savefig(dirname + "features.png")

    return clf, labels_test, predicted
```

```python
def display_predvstruth(
    feature_columns, continuous_col, target_column, sample_df, query, clf0to30
):
    datachart = sample_df[
        [
            "first_movement_start",
            "avg_delay_30min",
            "avg_delay_in30min",
            "avg_delay_in30min_cat",
            "avg_delay_in60min",
            "avg_delay_in60min_cat",
        ]
    ]
    next_delay_chart = (
        alt.Chart(datachart)
        .mark_line()
        .encode(x="first_movement_start:T", y=continuous_col)
    )

    lower_th_chart = (
        alt.Chart(datachart)
        .mark_rule(color="green")
        .encode(y="y_min:Q")
        .transform_calculate(y_min="4")
    )
    higher_th_chart = (
        alt.Chart(datachart)
        .mark_rule(color="orange")
        .encode(y="y_max:Q")
        .transform_calculate(y_max="6.5")
    )
    next_delay_andth_chart = next_delay_chart + lower_th_chart + higher_th_chart

    flight_df = sample_df.query(query)[
        feature_columns + [target_column, "first_movement_start"]
    ].dropna()
    data_df = flight_df[feature_columns + [target_column]]
    input_df = data_df[feature_columns]
    target_df = data_df[[target_column]]

    predicted = clf0to30.predict(input_df)

    res_comp_df = pd.DataFrame(
        {
            "time": flight_df.first_movement_start,
            "truth": flight_df[target_column].values,
            "pred": predicted,
        }
    )

    predVStruth_chart = (
        alt.Chart(res_comp_df)
        .mark_area(opacity=0.3)
        .transform_fold(["truth", "pred"])
        .encode(
            x="time:T",
            y=alt.Y("value:Q", stack=False),
            color="key:N",
            tooltip=["value:Q", "key:N"],
        )
    )
    return (next_delay_andth_chart & predVStruth_chart).resolve_scale(y="independent")


sample_df = dep_arr_df.query(
    "mvt_type=='DEP' & '2019-10-01 10:00:00+00:00'<first_movement_start<='2019-10-01 20:00:00+00:00'"
)
feature_columns = all_no_na_cols_0
target_column = "avg_delay_in60min_cat"
continuous_col = "avg_delay_in60min"
query = "parking_position==parking_position"

display_predvstruth(
    feature_columns, continuous_col, target_column, sample_df, query, clf30to60
)
```
